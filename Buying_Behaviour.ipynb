# Performing clustering according to the price of the vehicle owned

```python
data_price = data.drop('Make',axis=1)
```

```python
data_price
```

```python
price_data = data_price[['Price']].values.reshape(-1, 1)
```

```python
from sklearn.cluster import KMeans
cluster = range(1,11)
sse = []
for k in cluster:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(price_data)
    sse.append(kmeans.inertia_)

# Plot the SSE for each k
plt.figure(figsize=(10, 6))
plt.plot(cluster, sse, marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('SSE')
plt.title('Elbow Method for Determining Optimal Number of Clusters')
plt.show()
```

```python
# Choose based on the elbow plot
kmeans = KMeans(n_clusters=3, random_state=42)
model = kmeans.fit(price_data)
data_price['Price Cluster'] = model.labels_

# Display the first few rows with the cluster assignments
data_price[['Price' , 'Price Cluster']].head()                            
```

```python
# Visualize the clusters with a scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x=data.index, y='Price', hue='Price Cluster', data=data_price, palette='Set1')
plt.xlabel('Index')
plt.ylabel('Price')
plt.title('Clustering of Vehicles by Price')
plt.legend(title='Price Cluster')
plt.show()
```

```python
d2 = data_price[data_price['Price Cluster']==2]
d1 = data_price[data_price['Price Cluster']==1]
d0 = data_price[data_price['Price Cluster']==0]

plt.scatter(d1.Price, d1['Total Salary'],color='green')
plt.scatter(d0.Price, d0['Total Salary'],color='red')
plt.scatter(d2.Price, d2['Total Salary'],color='blue')
plt.xlabel('Price')
plt.ylabel('Total salary')
```

# Performing clustering according to the make of the vehicle

```python
data_make = data.drop('Price', axis=1)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
cols = ['Make']
data_make[cols] = data_make[cols].apply(le.fit_transform)
```

```python
data_make
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
x = data_make['Salary'].values.reshape(-1, 1)
x_scaled = scaler.fit_transform(x)
data_make['Salary']= x_scaled


x = data_make['Age'].values.reshape(-1, 1)
scaler = MinMaxScaler()
x_scaled = scaler.fit_transform(x)
data_make['Age']= x_scaled


x = data_make['Wife Salary'].values.reshape(-1, 1)
x_scaled = scaler.fit_transform(x)
data_make['Wife Salary']= x_scaled


x = data_make['Total Salary'].values.reshape(-1, 1)
x_scaled = scaler.fit_transform(x)
data_make['Total Salary']= x_scaled


```

```python
data_make
```

```python
cluster = range(1,11)
wcss = []
for i in cluster:
    model = KMeans(n_clusters=i, random_state=42)
    model.fit(data_make)
    wcss.append(model.inertia_)
    
plt.figure(figsize=(10, 6))
plt.plot(cluster, wcss, marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.title('Elbow Method for Determining Optimal Number of Clusters')
plt.show()
```

```python
kmeans = KMeans(n_clusters=3, random_state=42)
model = kmeans.fit(data_make)
data_make['Make Cluster'] = model.labels_

# Display the first few rows with the cluster assignments
data_make[['Make' , 'Make Cluster']].head()                            
```

```python
# Visualize the clusters with a scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x=data_make.index, y='Make', hue='Make Cluster', data=data_make, palette='Set1')
plt.xlabel('Index')
plt.ylabel('Make')
plt.title('Clustering of Vehicles by Make')
plt.legend(title='Make Cluster')
plt.show()
```
